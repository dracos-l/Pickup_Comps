```python
counter_dic = {}

for a in range(0,38):
    counter = 0
    for b in range(0,351):
        if float(basketball[headers[a]].iloc[b]) == 0:
            counter += 1
    counter_dic[headers[a]] = counter
```

```python
basketball.loc[basketball["Diff%_6"] == 0.0, "Diff%_6"] = 100.0
basketball.loc[basketball["Diff%_15"] == 0.0, "Diff%_15"] = 100.0
basketball.loc[basketball["eFG%_Iso_Def"] == 0.0, "eFG%_Iso_Def"] = 100.0
```

```python

the_master_dic = {}
for j in range(0,38):
    if headers[j] == 'Diff%_6' or headers[j] == 'Diff%_15' or headers[j] == 'eFG%_Iso_Def':
        df = basketball.sort_values(headers[j], ascending=False)
    else:
        df = basketball.sort_values(headers[j], ascending=True)
    for i in range(0, counter_dic[headers[j]]):
        the_master_dic[str(df['Player'].iloc[i])].append([headers[j], round(((counter_dic[headers[j]]/2)/350)*100)])
    for i in range(counter_dic[headers[j]],351):
        if headers[j] != 'Height':            
            the_master_dic[str(df['Player'].iloc[i])].append([headers[j], round((i/350)*100)])
        else:
            the_master_dic[str(basketball['Player'].iloc[i])] = []
            the_master_dic[str(basketball['Player'].iloc[i])].append([headers[j], int(float(basketball['Height'].iloc[i]))])

```

```python
the_master_list = []

for k in range(0,351):
    the_master_list.append([str(basketball['Player'].iloc[k])])
    for l in range (0,38):
        the_master_list[k].append(the_master_dic[str(basketball['Player'].iloc[k])][l][1])
the_real_master_list = []
the_real_master_list.append(headers_with_player)
for m in range(0,351):
    the_real_master_list.append(the_master_list[m])
print(the_real_master_list)

np.savetxt("Final_Final_Player_Precentile.csv", the_real_master_list, delimiter=", ", fmt="% s")
```
