```python
person = []
ast_precent = {}

def Average(lst):
    if len(lst) == 0:
        return 0
    return sum(lst) / len(lst)

for k in range(0,351):
    for i in range(0,1000):
        df = basketball[(basketball['%AST'] >= basketball['%AST'].quantile(i/1000)) & (basketball['%AST'] <= basketball['%AST'].quantile((i+1)/1000))]
        if df.empty:
            pass
        for j in range(0, len(df['Player'])):
            if df['Player'].iloc[j] == basketball['Player'].iloc[k]:
                person.append(i/10)

    ast_precent[str(basketball['Player'].iloc[k])] = round(Average(person)) #dictionary of everyone's precentile
    person = []
    print(ast_precent)

```
